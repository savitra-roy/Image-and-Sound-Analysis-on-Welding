```python
df= pd.read_csv('result.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df1 = pd.read_excel('data.xlsx', nrows=28, usecols=range(7))
df1.isna().sum()
```

```python
df1['rms'].fillna(df1['rms'].mean(), inplace=True)
df1.drop(columns=df1.columns[0], inplace=True)
df1['hbyw']= df1['height(mm)']/ df1['widht(mm)']
df1.drop(columns=['height(mm)', 'widht(mm)'], inplace=True)
X = df1['hbyw'].apply(lambda x: 1 if 0.3 <= x <= 0.7 else 0)
```

```python
X.drop([5,17],inplace=True)
```

```python
X.reset_index(drop=True, inplace=True)
```

```python
X
```

```python
df['weld']=X
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
df.columns
```

```python
X= df.drop(columns=['experiment', 'weld'])
y= df['weld']
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
model= SVC(kernel='rbf', gamma=0.1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
print(f'testing accuracy:{accuracy_score(y_pred,y_test)}')
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(f'training accuracy:{accuracy_score(model.predict(X_train),y_train)}')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

scores = cross_val_score(SVC(gamma=0.1), X, y, cv=5)
print("Cross-Validation Accuracy:", scores)
print("Average Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
scores = cross_val_score(SVC( gamma=0.1), X, y, cv=5)
print("LOOCV Accuracy:", scores)
print("LOOCV Accuracy:", scores.mean())
```
